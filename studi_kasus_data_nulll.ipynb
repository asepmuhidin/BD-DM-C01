```python
import pandas as pd
ffile="https://raw.githubusercontent.com/asepmuhidin/BD-DM-C02/refs/heads/main/Tabel_Penjualan_Rupiah.csv"

dataset=pd.read_csv(ffile)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset['Harga (Rupiah)'] = dataset['Harga (Rupiah)'].str.replace(',', '', regex=False).astype(float)
```

```python
dataset.info()
```

```python
dataset.fillna({
    'Jumlah':dataset['Jumlah'].mean(),
    'Harga (Rupiah)':dataset['Harga (Rupiah)'].mean(),
},inplace=True)
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)
dataset['Total Harga (Rupiah)'] = dataset['Jumlah'] * dataset['Harga (Rupiah)']
dataset.head(10)
```
